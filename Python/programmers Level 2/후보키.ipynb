### >> 문제: programmers L2 [후보키](https://programmers.co.kr/learn/courses/30/lessons/42890)

프렌즈대학교 컴퓨터공학과 조교인 제이지는 학생들의 인적사항을 정리하는 업무를 담당하게 되었다.  

모든 인적사항을 데이터베이스에 넣기로 하였고,   
이를 위해 정리를 하던 중에 후보키(Candidate Key)에 대한 고민이 필요하게 되었다.  

관계 데이터베이스에서 릴레이션(Relation)의 튜플(Tuple)을  
유일하게 식별할 수 있는 속성(Attribute) 또는 속성의 집합 중,  
다음 두 성질을 만족하는 것을 후보 키(Candidate Key)라고 한다.  

* 유일성(uniqueness) : 릴레이션에 있는 모든 튜플에 대해 유일하게 식별되어야 한다.  
* 최소성(minimality) : 유일성을 가진 키를 구성하는 속성(Attribute) 중   
하나라도 제외하는 경우 유일성이 깨지는 것을 의미한다.   
즉, 릴레이션의 모든 튜플을 유일하게 식별하는 데 꼭 필요한 속성들로만 구성되어야 한다.  

제이지를 위해, 아래와 같은 학생들의 인적사항이 주어졌을 때, 후보 키의 최대 개수를 구하라.

위의 예를 설명하면, 학생의 인적사항 릴레이션에서 모든 학생은 각자 유일한 학번을 가지고 있다.  
따라서 학번은 릴레이션의 후보 키가 될 수 있다.  

그다음 이름에 대해서는 같은 이름(apeach)을 사용하는 학생이 있기 때문에, 이름은 후보 키가 될 수 없다.  
그러나, 만약 [이름, 전공]을 함께 사용한다면 릴레이션의 모든 튜플을 유일하게 식별 가능하므로 후보 키가 될 수 있게 된다.  
물론 [이름, 전공, 학년]을 함께 사용해도 릴레이션의 모든 튜플을 유일하게 식별할 수 있지만,  
최소성을 만족하지 못하기 때문에 후보 키가 될 수 없다.  
따라서, 위의 학생 인적사항의 후보키는 학번, [이름, 전공] 두 개가 된다.  

릴레이션을 나타내는 문자열 배열 relation이 매개변수로 주어질 때,  
이 릴레이션에서 후보 키의 개수를 return 하도록 solution 함수를 완성하라.  

[제한사항]
* relation은 2차원 문자열 배열이다.
* relation의 컬럼(column)의 길이는 1 이상 8 이하이며, 각각의 컬럼은 릴레이션의 속성을 나타낸다.
* relation의 로우(row)의 길이는 1 이상 20 이하이며, 각각의 로우는 릴레이션의 튜플을 나타낸다.
* relation의 모든 문자열의 길이는 1 이상 8 이하이며, 알파벳 소문자와 숫자로만 이루어져 있다.
* relation의 모든 튜플은 유일하게 식별 가능하다.(즉, 중복되는 튜플은 없다.)

*출처: 프로그래머스 코딩 테스트 연습, https://programmers.co.kr/learn/challenges*

### >> 풀이
모든 조합을 구하여 유일성을 판별한다

```python
from itertools import combinations

def solution(relation):
    answer = 0
    
    # 가능한 모든 조합
    temp = [[] for i in range(len(relation[0]))]
    for row in relation:
        for i in range(1,len(row)+1):
            temp[i-1].append(list(combinations(row,i)))
#     print('combinations',temp,"\n")

    # superkey를 만들기 위한 데이터 전처리
    superkey = [[[] for j in range(len(temp[i][0]))] for i in range(len(temp[0][0]))]
    for i in range(len(temp)):
        for j in range(len(temp[i])):
            for k in range(len(temp[i][j])):
                superkey[i][k].append(temp[i][j][k])
#     print('superkey',superkey,'\n')

    # 유일성을 만족하지만, 최소성을 만족하지 못하는 superkey를 candidate에 담는다
    candidate = [[] for i in range(len(relation[0]))]
    for i in range(len(superkey)):
        for j in range(len(superkey[i])):
            if len(superkey[i][j]) == len(set(superkey[i][j])):
                candidate[i].append(superkey[i][j][0])
#     print('candidate',candidate,'\n')

    # 후보키 찾기 !!
    left, right = 0, 1
    while right < len(candidate):  
        for i in range(len(candidate[left])):
            j = 0
            while j < len(candidate[right]):
                if not set(candidate[left][i]) - set(candidate[right][j]):
                    candidate[right].remove(candidate[right][j])
                else:
                    j += 1
        right += 1
        if right == len(candidate):
            left += 1
            right = (left+1)
#     print('candidate',candidate,"\n")
    
    # 후보키 개수 구하기
    for i in candidate:
        answer += len(i)
    
    return answer

print(solution([["100","ryan","music","2"],
                ["200","apeach","math","2"],
                ["300","tube","computer","3"],
                ["400","con","computer","4"],
                ["500","muzi","music","3"],
                ["600","apeach","music","2"]])," >> 2\n")

print(solution([["ryan","100","music","2"],
                ["apeach","200","math","2"],
                ["tube","300","computer","3"],
                ["con","400","computer","4"],
                ["muzi","500","music","3"],
                ["apeach","600","music","2"]])," >> 2\n")

print(solution([["a","1","4"],["2","1","5"],["a","2","4"]])," >> 2\n")
print(solution([["a","b","c"],["1","b","c"],["a","b","4"],["a","5","c"]])," >> 1\n")
print(solution([['1','1','3']]), " >> 3\n")
print(solution([['1']]), " >> 1\n")
print(solution([['1'],['1']]), " >> 0\n")
```

### >> 다른풀이

모든 조합의 **인덱스**를 구한 후, 인덱스에 해당되는 값으로 유일성을 판별한다

```python
from itertools import combinations

# 유일성을 만족하는 슈퍼키 찾기
def isSuperkey(combi, relation):
    # relation의 행 중에서 combi(인덱스)에 해당하는 내용을 튜플로 담는다 
    temp = [tuple([row[i] for i in combi]) for row in relation]
    if len(set(temp)) == len(temp): # 유일하게 식별 가능하면 True
        return True 
    return False

def solution(relation):    
    candidate = []
    
    # 슈퍼키인 인덱스 조합 찾기
    for i in range(1,len(relation[0])+1): # 열의 수 만큼 반복
        # 열 인덱스의 모든 조합 
        index_combination = list(combinations(range(len(relation[0])),i))
        # 열 인덱스의 모든 조합 중 슈퍼키 찾기
        for combi in index_combination:
            if isSuperkey(combi, relation): # 슈퍼키이면 
                candidate.append(set(combi)) # 인덱스 조합을 candidate에 담는다

    # 후보키 찾기
    for i in candidate[:]:
        for j in candidate[:]:
            if not i-j and i!=j:
                candidate.remove(j)

    return len(candidate)

print(solution([["100","ryan","music","2"],
                ["200","apeach","math","2"],
                ["300","tube","computer","3"],
                ["400","con","computer","4"],
                ["500","muzi","music","3"],
                ["600","apeach","music","2"]])," >> 2\n")

print(solution([["ryan","100","music","2"],
                ["apeach","200","math","2"],
                ["tube","300","computer","3"],
                ["con","400","computer","4"],
                ["muzi","500","music","3"],
                ["apeach","600","music","2"]])," >> 2\n")

print(solution([["a","1","4"],["2","1","5"],["a","2","4"]])," >> 2\n")
print(solution([["a","b","c"],["1","b","c"],["a","b","4"],["a","5","c"]])," >> 1\n")
print(solution([['1','1','3']]), " >> 3\n")
print(solution([['1']]), " >> 1\n")
print(solution([['1'],['1']]), " >> 0\n")
```
